```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn import tree
import matplotlib.pyplot as plt

# Load the Iris dataset (or you can use your own dataset)
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
class_names = iris.target_names

# Create a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Visualize the decision tree
fig, ax = plt.subplots(figsize=(12, 10))
tree.plot_tree(clf, feature_names=feature_names, class_names=class_names, filled=True, ax=ax)
plt.show()
```

```python
new_data = [[5.1, 3.5, 1.4, 0.2]]  # Example new data point
predicted_class = clf.predict(new_data)
print("Predicted class:", class_names[predicted_class[0]])
```
